```python
import re

query = []

def GetTheStatements():
    with open('sample_stored_procedure.sql') as fp:
        for result in re.findall('(.*?);', fp.read(), re.S):
            query.append(result)
    print(query)

GetTheStatements()
```

```python
import re
for strval in query:
    if "INSERT INTO " in strval:
        d = {}
        c += 1
        d["statement_id"] = c
        d["statement_type"] = "insert"
        target = []
        if "FROM " in strval:
            Reg = re.compile(r'(?<=FROM )\w+.\w+')
            mo1 = Reg.findall(strval)
            s = set(mo1)
            for val in s:
                dtemp = {"type" : "from","name" : val}
            target.append(dtemp)
        if "JOIN " in strval:
            Reg = re.compile(r'(?<=JOIN )\w+.\w+')
            mo1 = Reg.findall(strval)
            s = set(mo1)
            for val in s:
                dtemp = {"type" : "join","name" : val}
            target.append(dtemp)
            
        Regex = re.compile(r'(?<=INSERT INTO )\w+')
        mo1 = Regex.search(strval)
        d[mo1.group(0)] = target
        dfinal.append(d)
    if "UPDATE " in strval:                         
        d = {}
        c += 1
        d["statement_id"] = c
        d["statement_type"] = "update"
        target = []
        if "FROM " in strval:
            Reg = re.compile(r'(?<=FROM )\w+.\w+')
            mo1 = Reg.findall(strval)
            s = set(mo1)
            for val in s:
                dtemp = {"type" : "from","name" : val}
            target.append(dtemp)
        if "JOIN " in strval:
            Reg = re.compile(r'(?<=JOIN )\w+.\w+')
            mo1 = Reg.findall(strval)
            s = set(mo1)
            for val in s:
                dtemp = {"type" : "join","name" : val}
            target.append(dtemp)
            
        Regex = re.compile(r'(?<=UPDATE )\w+')
        mo1 = Regex.search(strval)
        d[mo1.group(0)] = target
        dfinal.append(d)
    if "CREATE OR REPLACE TABLE " in strval:
        d = {}
        c += 1
        d["statement_id"] = c
        d["statement_type"] = "create"
        target = []
        if "FROM " in strval:
            Reg = re.compile(r'(?<=FROM )\w+.\w+')
            mo1 = Reg.findall(strval)
            s = set(mo1)
            for val in s:
                dtemp = {"type" : "from","name" : val}
            target.append(dtemp)
        if "JOIN " in strval:
            Reg = re.compile(r'(?<=JOIN )\w+.\w+')
            mo1 = Reg.findall(strval)
            s = set(mo1)
            for val in s:
                dtemp = {"type" : "join","name" : val}
            target.append(dtemp)
            
        Regex = re.compile(r'(?<=CREATE OR REPLACE TABLE )\w+')
        mo1 = Regex.search(strval)
        d[mo1.group(0)] = target
        dfinal.append(d)
    if "DELETE FROM " in strval:
        d = {}
        c += 1
        d["statement_id"] = c
        d["statement_type"] = "delete"
        Regex = re.compile(r'(?<=DELETE FROM )\w+')
        mo1 = Regex.search(strval)
        d[mo1.group(0)] = []
        dfinal.append(d)
        
```
